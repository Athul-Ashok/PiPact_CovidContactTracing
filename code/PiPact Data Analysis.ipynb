```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
basePath = "/Users/athulashok/"
advin_microwave_3ft = pd.read_csv(os.path.join(basePath,"advin_microwave_3ft.csv"))
microwave_off_3ft = pd.read_csv (os.path.join(basePath,"microwave_off_3ft.csv"))
microwave_on_3ft = pd.read_csv (os.path.join(basePath,"microwave_on_3ft.csv"))
no_microwave_3ft = pd.read_csv (os.path.join(basePath,"no_microwave_3ft.csv"))
fluctuating3_9ft = pd.read_csv (os.path.join(basePath,"fluctuating3_9ft.csv"))
humanwalk_acrossroom6_12ft = pd.read_csv (os.path.join(basePath,"humanwalk_acrossroom6_12ft.csv"))
twoscanners_data_one = pd.read_csv (os.path.join(basePath,"twoscanners_data_one.csv"))
twoscanners_data_two = pd.read_csv (os.path.join(basePath,"twoscanners_data_two.csv"))
```

```python
def plotRSSI(beaconDataFrame,title):
    rssi_mean = [np.mean(beaconDataFrame['RSSI'])]*beaconDataFrame['RSSI'].size
    fig,ax = plt.subplots()
    ax.plot(beaconDataFrame['SCAN'],beaconDataFrame['RSSI'], label='RSSI')
    # Plot the average line
    ax.plot(beaconDataFrame['SCAN'],rssi_mean, label='Mean', linestyle='--')
    # Make a legend
    legend = ax.legend(loc='lower right')
    plt.title(title) 
```

```python
df_1= DataFrame(no_microwave_3ft,columns=(['RSSI']))
axes = df_1.plot()
df_2= DataFrame(microwave_off_3ft,columns=(['RSSI']))
df_2.plot(ax=axes)
df_3= DataFrame(microwave_on_3ft,columns=(['RSSI']))
df_3.plot(ax=axes)
df_4= DataFrame(advin_microwave_3ft,columns=(['RSSI']))
df_4.plot(ax=axes)
plt.suptitle('Microwave Experiment Summative Data')
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
```

```python
plotRSSI(no_microwave_3ft,"Beacons with No Microwave in Between - 3ft no MW")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
plotRSSI(microwave_off_3ft,"Beacons with Microwave in Between - 3ft MW OFF")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
plotRSSI(microwave_on_3ft,"Beacons with Microwave in Between - 3ft MW ON")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
plotRSSI(advin_microwave_3ft,"Beacons with Advertiser in  Microwave - 3ft MW OFF")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
```

```python
df_1= DataFrame(twoscanners_data_one,columns=(['RSSI']))
axes = df_1.plot()
df_2= DataFrame(twoscanners_data_two,columns=(['RSSI']))
df_2.plot(ax=axes)
plt.suptitle('Two Scanners Summative Data')
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
```

```python
plotRSSI(twoscanners_data_one,"Two Scanners Dataset One")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
plotRSSI(twoscanners_data_two,"Two Scanners Dataset Two")
plt.xlabel('Time(Seconds)')
plt.ylabel('RSSI')
```
